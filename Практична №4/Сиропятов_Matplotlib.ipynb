## Лінійні графіки

```python
import matplotlib.pyplot as plt
import pandas as pd

date = pd.date_range(start='2022-01-01', freq='D', periods=8)

# 1
# f = plt.figure()
# f.set_figwidth(10)
# f.set_figheight(8)

# 2
# plt.figure(figsize=(10, 8))

# 3
plt.rcParams['figure.figsize'] = [10, 8]

plt.plot(date, [-1, -2, -5, 0, -4, -3, 0, -4])
plt.show()
```

```python
# matplotlib.text.Text
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')

plt.title('Середньодобова температура за день в місті')

plt.text(date[2], -5, 'Найпрохолодніший день')
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
plt.plot(date, [-1, -2, -5, 0, -4, -3, 0, -4], label='mean day temperature')
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Середньодобова температура за день в місті')
plt.text(date[2], -5, 'Найпрохолодніший день')
plt.legend()
plt.show()
```

## Компонування графіків

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
fig, axs = plt.subplots()
axs.plot(date, [-1, -2, -5, 0, -4, -3, 0, -4], label='mean day temperature')
axs.plot(date, [-5, -6, -8, -3, -5, -7, -1, -3], label='mean night temperature')
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Середньодобова температура за день в місті')
plt.text(date[2], -5.2, 'Найпрохолодніша температура вдень')
plt.text(date[2], -8.2, 'Найпрохолодніша температура вночі')
plt.legend()
plt.show()
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
fig, axs = plt.subplots(2, 1)
axs[0].plot(date, [-1, -2, -5, 0, -4, -3, 0, -4], label='mean day temperature')
axs[1].plot(date, [-5, -6, -8, -3, -5, -7, -1, -3], label='mean night temperature')

axs[0].set_title('Вдень', fontsize=10)
axs[1].set_title('Вночі', fontsize=10)

fig.suptitle('Середньодобова температура за день в місті')

plt.show()
```

## Налаштування графіків


```python
## linestyle
# - (solid)
# -- (dashed)
# -. (dashplot)
# : (dotted)


## color (c)
# (0.5, 0.6, 0.7)
# #0a0a0a
# b, g, r, c, m, y, k, w
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
plt.plot(
    date,
    [-1, -2, -5, 0, -4, -3, 0, -4],
    label="mean day temperature",
    linestyle="-",
    color="#FF5733"
)
plt.plot(
    date,
    [-5, -6, -8, -3, -5, -7, -1, -3],
    label="mean night temperature",
    linestyle="-.",
    color="#061358"
)
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Середньодобова температура за день в місті')
plt.legend()
plt.show()
```

```python
## marker
# .
# o
# +
# x
# D (ромб)
# v (трикутник з вершиною вниз)
# ^ (трикутник з вершиною вверх)
# s (квадрат)
# * (п'ятикутна зірка)
```

```python
date = pd.date_range(start='2022-01-01', freq='D', periods=8)
plt.plot(
    date,
    [-1, -2, -5, 0, -4, -3, 0, -4],
    label="mean day temperature",
    linestyle="--",
    color="#FF5733",
    linewidth=2,
    marker="D"
)
plt.plot(
    date,
    [-5, -6, -8, -3, -5, -7, -1, -3],
    label="mean night temperature",
    linestyle="-.",
    color="#061358",
    linewidth=2,
    marker="^"
)
plt.ylim(-9, 1)
plt.xlabel('Дата', fontsize='small', color='midnightblue')
plt.ylabel('Температура', fontsize='small', color='midnightblue')
plt.title('Середньодобова температура за день в місті')
plt.legend()
plt.grid()
plt.show()
```

# Стовпчикові діаграми
width - ширина\
bottom - нижні кордони стовпців\
height - послідовність значень висоти\
color - колір заливки\
edgecolor - колір меж стовпців\
linewidth - значення товщини ліній меж стовпців (в `pt`)\
xerr, yerr - граничні значення для діаграм похибок\
align - вирівнювання стовпців


```python
plt.bar(
    ['USA', 'China', 'Japan'],
    [10, 15, 20],
    color=['b', 'r', 'y'],
    edgecolor='g',
    )
plt.show()
```

```python
import numpy as np

plt.rcdefaults()

fig, ax = plt.subplots()

people = ('Tom', 'Harry', 'Jim')
y_pos = np.arange(len(people))
perfomance = 3 + 10 * np.random.rand(len(people))
error = np.random.rand(len(people))

ax.barh(y_pos, perfomance, xerr=error, align='center')
ax.set_yticks(y_pos, labels=people)
ax.invert_yaxis()

ax.set_xlabel('Perfomance')
ax.set_title('Як швидко Ви хочете піти з офісу?')

plt.show()
```

```python
labels = ['Junior', 'Senior', 'Engineer', 'SysAdmin']
data = [64, 32, 100, 2]
explode = [0, 0, 0, 0]

plt.pie(
    data,
    labels=labels,
    shadow=True,
    explode=explode,
    autopct="%.2f%%",
    pctdistance=1.15,
    labeldistance=1.35
)
plt.show()
```

```python
theta = np.linspace(0, 2.0 * np.pi, 1000)

r = np.sin(6 * theta)
plt.polar(theta, r);
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

theta_max = 8 * np.pi
n =1000
theta = np.linspace(0, theta_max, n)
x = theta
z = np.sin(theta)
y = np.cos(theta)
ax.plot(x, y, z, 'g');
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = [5, 10, 15, 20]
z = [10, 0, 5, 15]
y = [0, 10, 5, 25]
s = [150, 130, 30, 160]
ax.scatter(x, y, z, s=s)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_wireframe(x, y, z)

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")

grid = np.arange(-10, 10, 0.5)
x, y = np.meshgrid(grid, grid)
z = x ** 2 * y ** 2 + 2

ax.plot_surface(x, y, z)

plt.show()
```
